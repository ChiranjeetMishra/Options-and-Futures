## Payoff function

### Call Option
<center><I>payoff=max(-C, S - X - C)</I></center>

### Put Option
<center><I>payoff=max(-P, X - S - P)</I></center>

P and C are premiums of calls and puts<br/>
S is the stock price<br/>
X is the strike price

```python
def payoff(option_type, premium, S, X):
    """
    Calculates the payoff for call and put options.
    
    Parameters
    ----------
    option_type : char
        C for call
        P for put
        
    premium : float
        the premium of call or put option
        
    S : float
        Stock Price
        
    X : float
        Strike Price
    
    Returns
    -------
    Payoff : float
        The payoff of call and put options
    """
    Payoff = 0
    if option_type == 'C':
        Payoff = max(-premium, S-X-premium)
    elif option_type == 'P':
        Payoff = max(-premium, X-S-premium)
    else:
        print("Option type not supported")
    return Payoff
```

```python
payoff('C', 10, 120,100)  #answer displayed should be 10
```

```python
payoff('Z', 10, 120, 100)  #answer displayed should be option type not supported
```

```python
payoff('P', 10, 120, 100)  #answer displayed should be -10
```

## Payoff Graphs

### Call Option

```python
X = dfo_exp.calls['strike'][0]
premium = dfo_exp.calls['lastPrice'][0]
```

```python
StockPrice = np.arange(0, 200)
Payoff = [payoff('C', premium, i, X) for i in StockPrice]
```

```python
plt.plot(StockPrice, Payoff, label = 'payoff line')
plt.xlabel('Stock Price')
plt.ylabel('payoff')
plt.title('Call option Payoff at Expiry')
plt.grid(True)
plt.plot(X+premium, 0, 'ro', label = 'breakeven point')
plt.legend()
```

### Put Option

```python
X = dfo_exp.puts['strike'][len(dfo_exp.puts)-1]
premium = dfo_exp.puts['lastPrice'][len(dfo_exp.puts)-1]
```

```python
StockPrice = np.arange(0, 300)
Payoff = [payoff('P', premium, i, X) for i in StockPrice]
```

```python
plt.plot(StockPrice, Payoff, label = 'payoff line')
plt.xlabel('Stock Price')
plt.ylabel('payoff')
plt.title('Call option Payoff at Expiry')
plt.grid(True)
plt.plot(X-premium, 0, 'ro', label = 'breakeven point')
plt.legend()
```
